```r
# Charger les données
sabra <- read.csv("C:/Users/Mahdi-ZIANE/OneDrive/pluie_obs/processed/sabra160502.csv")
Bensekrane <- read.csv("C:/Users/Mahdi-ZIANE/OneDrive/pluie_obs/processed/2160702_Bensekrane.csv")
Hennaia <- read.csv("C:/Users/Mahdi-ZIANE/OneDrive/pluie_obs/processed/160516 M lala Hennaia.csv")

# Calculer les moyennes
sabra_mean <- mean(sabra$annuel, na.rm=TRUE)
Bensekrane_mean <- mean(Bensekrane$annuel, na.rm=TRUE) 
Hennaia_mean <- mean(Hennaia$annuel, na.rm=TRUE)

# Créer le graphique 
par(mfrow=c(1,1))
boxplot(sabra$annuel, col="black", ylim=c(0,170), outline=FALSE,
        boxwex=0.6, staplewex=0.6, medlwd=2, medcol="gray")
boxplot(Bensekrane$annuel, col="blue", add=TRUE, 
        boxwex=0.6, staplewex=0.6, medlwd=3, medcol="lightblue")
boxplot(Hennaia$annuel, col="red", add=TRUE,
        boxwex=0.6, staplewex=0.6, medlwd=4, medcol="pink")

# Ajouter légende
legend("topright", legend=c("Sabra","Bensekrane","Hennaia"),
       col=c("black","blue","red"), lwd=2)
       
# Ajouter moyennes       
points(1, sabra_mean, pch=4, col="black", cex=2)
points(2, Bensekrane_mean, pch=4, col="blue", cex=2)
points(3, Hennaia_mean, pch=4, col="red", cex=2)

# Titres axes
title("Boxplots des séries")
```

```r
# Load libraries
library(ggplot2)

# Read in data
obs <- read.csv("C:/Users/Mahdi-ZIANE/OneDrive/pluie_obs/processed_Precip_Record/observed/160705_lala_setti.csv")
persian <- read.csv("C:/Users/Mahdi-ZIANE/OneDrive/Persian_cdr_Pjmax/LALA SETTI_annual_max.csv") 
cmip6 <- read.csv("C:/Users/Mahdi-ZIANE/OneDrive/pluie_obs/processed_Precip_Record/from_Models/pr_day_CanESM5_historical_r1i1p1f1_gn_19500101-20141231_v20190429_LALA SETTI.csv")

# Extract rainfall columns
obs_rain <- obs$annuel
persian_rain <- persian$precip
cmip6_rain <- cmip6$pr

# Plot boxplots
ggplot(data = NULL) +
  geom_boxplot(data = obs_rain, aes(x = "", y = obs_rain), color = "black") +
  geom_boxplot(data = persian_rain, aes(x = "", y = persian_rain), color = "blue") +
  geom_boxplot(data = cmip6_rain, aes(x = "", y = cmip6_rain), color = "red") +
  scale_x_discrete(name = "Source", labels = c("Observed", "PERSIAN", "CMIP6")) +
  
  # Add mean markers
  geom_point(aes(x = 1, y = mean(obs_rain)), color="black", size=3, shape=4) +
  geom_point(aes(x = 2, y = mean(persian_rain)), color="blue", size=3, shape=4) +
  geom_point(aes(x = 3, y = mean(cmip6_rain)), color="red", size=3, shape=4)
```
